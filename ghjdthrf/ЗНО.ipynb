```python
# функция для заполнения столбца с Федеральным округом.
def fo (region):
    if region in central_fo:
        return 'Центральный'
    if region in nord_west_fo:
        return 'Северо-Западный'
    if region in south_fo:
        return 'Южный'
    if region in nord_kavkaz_fo:
        return 'Северо-Кавказский'
    if region in privolzh_fo:
        return 'Приволжский'
    if region in ural_fo:
        return 'Уральский'
    if region in siberia_fo:
        return 'Сибиркий'
    if region in dalnevostok_fo:
        return 'Северо-Западный'    
```

```python
# с помощью цикла создадим новый столбец во всех таблицах и заполним его соответствующим Федеральным округом.
for df in full_list_of_tables:
    df['Федеральный округ'] = df['Республика, край, область'].apply(fo)
```
